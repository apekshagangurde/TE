## 1. 

```python
import pandas as pd

data = pd.read_csv('D:/Programming Stuff/Programs/ML Files/Datasets/toy_dataset.csv')
data.head(10)
```

```python
# method 1
data.describe()
```

```python
# method 2
funs = ['mean','median','std','min','max']
data['Age'].apply(funs)
```

```python
data['Income'].apply(funs)
```

```python
mean = data.groupby('Age')['Income'].mean()
minimum = data.groupby('Age')['Income'].min()
maximum = data.groupby('Age')['Income'].max()
std = data.groupby('Age')['Income'].std()
med = data.groupby('Age')['Income'].median()
```

```python
pd.DataFrame({'mean':mean.values,
              'min':minimum.values,
              'maximum':maximum.values,
              'std':std.values,
              'med':med.values},
             index=mean.index)
```
