```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=["Age"], inplace = True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
fare_for_class = df.groupby(["Pclass"], as_index = False)["Fare"].sum()

sns.barplot(x="Pclass", y="Fare", data = fare_for_class)
plt.title("Fare for Passenger Class")
```

```python
sns.countplot(x="Pclass", data = df)
plt.title("Count of Passenger Class")
```

```python
survived = df[df["Survived"] == 1]

survived_for_class = survived.groupby(["Pclass", "Sex"], as_index = False)["Survived"].sum()

ax = sns.barplot(x="Pclass", y="Survived", data = survived_for_class, hue = "Sex")
plt.title("Passenger Class Survival Rate")
```

```python
sns.histplot(y="Fare", x="PassengerId", data=df)
plt.title("Fare distribution for each Passenger")
```
