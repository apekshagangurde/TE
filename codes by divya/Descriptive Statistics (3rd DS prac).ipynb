## 1. Provide summary statistics (mean, median, minimum, maximum, standard deviation) for a dataset (age, income etc.) with numeric variables grouped by one of the qualitative (categorical) variable. For example, if your categorical variable is age groups and quantitative variable is income, then provide summary statistics of income grouped by the age groups. ### Importing Data

```python
import pandas as pd
income = pd.read_csv("D:\Programming Stuff\Programs\ML Files\Datasets\income.csv")
```

```python
income.head()
```

### Summary Statistics

```python
grouped = income.groupby('Age_Group')['Income']
stats = pd.DataFrame({
    "Mean" : grouped.mean(),
    "Median" : grouped.median(),
    "Minimum" : grouped.min(),
    "Maximum" : grouped.max(),
    "Standard Deviation" : grouped.std()
})
stats
```

```python
income.describe()
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')

functions = ['mean','median','min','max','std']

for fun in functions:
    plt.plot(grouped.apply(fun),label=fun,marker='o')

plt.legend();
```

## 2. Write a Python program to display some basic statistical details like percentile, mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris- versicolor’ of iris.csv dataset.


```python
iris = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
```

```python
iris.head()
```

```python
species = iris.groupby('species')
quantile_25 = lambda x: x.quantile(0.25)
quantile_25.__name__ = '25th_percentile'
quantile_75 = lambda x: x.quantile(0.75)
quantile_75.__name__ = '75th_percentile'
details = species.agg(['mean','std',quantile_25,quantile_75])
details
```
